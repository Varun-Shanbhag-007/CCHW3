```python
import numpy as np
import matplotlib.pyplot as plt
 
plt.figure(figsize=(25,12))
plt.rcParams.update({'font.size': 15})
# set width of bar
barWidth = 0.4

plt.title("Benchmark Plot (Mean)",fontweight='bold',fontSize=40)
#our prg
bars1 = [402.9067619,402.9067619,441.0968095,403.4762381]

#iozone
bars2 = [409.069,409.069,445.4215238,408.1388571]

 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]

 
# Make the plot
plt.bar(r1, bars1,  width=barWidth, edgecolor='white', label='MyDisk BenchMark')
plt.bar(r2, bars2,  width=barWidth, edgecolor='white', label='Iozone')


plt.tick_params(axis='x', labelsize=15)
# Add xticks on the middle of the group bars
plt.xlabel('Operations', fontweight='bold')
plt.ylabel('MB/Sec', fontweight='bold')
plt.xticks([r + (barWidth/2)  for r in range(len(bars1))], ['Write Sequential','Read Sequential','Write Random','Read Random'],fontweight='bold')
 
# Create legend & Show graphic
plt.legend()
plt.show()
```
