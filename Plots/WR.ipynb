```python
import numpy as np
import matplotlib.pyplot as plt
 
plt.figure(figsize=(25,12))
plt.rcParams.update({'font.size': 15})
# set width of bar
barWidth = 0.4

plt.title("Write Random",fontweight='bold',fontSize=40)
#our prg
bars1 = [367.618,478.068,322.58,465.117,481.811,463.23,478.351,478.677,466.145,478.968,482.314,463.484,434,465,441,481,459,457,465,323,311.67]


#iozone
bars2 = [368.231,468,459.225,449,463.6,452.348,462,463.3,465.7,462.87,457,459.78,466,463.33,461.25,470.08,455,451,461.938,371.2,323]

 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]

 
# Make the plot
plt.bar(r1, bars1,  width=barWidth, edgecolor='white', label='MyDisk BenchMark')
plt.bar(r2, bars2,  width=barWidth, edgecolor='white', label='Iozone')


plt.tick_params(axis='x', labelsize=15)
# Add xticks on the middle of the group bars
plt.xlabel('Record  Size and Thhreads ', fontweight='bold')
plt.ylabel('MB/Sec', fontweight='bold')
plt.xticks([r + (barWidth/2)  for r in range(len(bars1))], ['64KB 1','1MB 1','16MB 1','64KB 2','1MB 2','16MB 2','64KB 4','1MB 4','16MB 4','64KB 8','1MB 8','16MB 8','64KB 12','1MB 12','16MB 12','64KB 24','1MB 24','16MB 24','64KB 48','1MB 48','16MB 48'], rotation=90)
 
# Create legend & Show graphic
plt.legend()
plt.show()
```
