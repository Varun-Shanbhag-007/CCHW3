```python
import numpy as np
import matplotlib.pyplot as plt
 
plt.figure(figsize=(25,12))
plt.rcParams.update({'font.size': 15})
# set width of bar
barWidth = 0.4

plt.title("Read Random IOPS",fontweight='bold',fontSize=40)
#our prg
bars1 = [27392,49152,33792,53504,69376,73984,85760]

#iozone
bars2 = [8518,16190,29513.6,50300.5,57301,65803.38,81441.3]

 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]

 
# Make the plot
plt.bar(r1, bars1,  width=barWidth, edgecolor='white', label='MyDiskBench')
plt.bar(r2, bars2,  width=barWidth, edgecolor='white', label='Iozone')


plt.tick_params(axis='x', labelsize=15)
# Add xticks on the middle of the group bars
plt.xlabel('Number of Threads', fontweight='bold')
plt.ylabel('OPS/Sec', fontweight='bold')
plt.xticks([r + (barWidth/2)  for r in range(len(bars1))], ['1','2','4','8','12','24','48'], rotation=0)

# Create legend & Show graphic
plt.legend()
plt.show()
```
